```python
# ==============================
# 1) IMPORTS
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
from IPython.display import display

# ==============================
# 2) LOAD FULL DATA FROM SQL
# ==============================
engine = create_engine(
    "mssql+pyodbc://LAPTOP-ENGNAGWA/SupplyChain"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&trusted_connection=yes"
    "&Encrypt=no"
)

# ‚úÖ ŸáŸÜÿß ÿ¨ÿ®ŸÜÿß ŸÉŸÑ ÿßŸÑÿ¨ÿØŸàŸÑ ŸÖŸÜ ÿ∫Ÿäÿ± TOP 10
df_original = pd.read_sql("SELECT * FROM SupplyChain", engine)

print("üìä Original dataset shape:", df_original.shape)
print("\nüìä Preview of dataset (first 5 rows):")
print(df_original.head())

# ==============================
# 3) BASIC INFO BEFORE CLEANING
# ==============================
print("\nüîπ Data Overview BEFORE cleaning:")
print(df_original.info())

print("\nüîπ Missing Values per Column (BEFORE):")
print(df_original.isnull().sum())

print("\nüîπ Number of duplicate rows (BEFORE):",
      df_original.duplicated().sum())

# ==============================
# 4) OUTLIERS BEFORE CLEANING (BOXPLOT)
# ==============================
def plot_outliers_histogram_and_boxplot(df_before, df_after):
    numeric_cols = df_before.select_dtypes(include='number').columns

    for col in numeric_cols:
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))

        # HISTOGRAM Before & After
        sns.histplot(df_before[col], color='red', label='Before Cleaning',
                     kde=True, bins=30, alpha=0.5, ax=axes[0])
        sns.histplot(df_after[col], color='green', label='After Cleaning',
                     kde=True, bins=30, alpha=0.5, ax=axes[0])
        axes[0].set_title(f"Distribution of '{col}' Before & After Cleaning")
        axes[0].set_xlabel(col)
        axes[0].legend()

        # BOX PLOT Before & After
        axes[1].boxplot(
            [df_before[col].dropna(), df_after[col].dropna()],
            labels=['Before', 'After'],
            patch_artist=True,
            boxprops=dict(facecolor='pink'),
            medianprops=dict(color='black')
        )
        axes[1].set_title(f"Boxplot of '{col}' Before & After Cleaning")
        axes[1].set_ylabel(col)

        plt.tight_layout()
        plt.show()


# ==============================
# 5) DATA CLEANING SECTION
# ==============================
print("\nüöø Starting data cleaning process...")

# ŸÜÿ¥ÿ™ÿ∫ŸÑ ÿπŸÑŸâ ŸÜÿ≥ÿÆÿ© ÿ¨ÿØŸäÿØÿ© ÿπŸÑÿ¥ÿßŸÜ ŸÜÿ≠ÿßŸÅÿ∏ ÿπŸÑŸâ ÿßŸÑÿ£ÿµŸÑŸäÿ© ŸÉŸÖÿß ŸáŸä
df_cleaned = df_original.copy()

# ---- 5.1 REMOVE DUPLICATES ----
rows_before = df_cleaned.shape[0]
df_cleaned = df_cleaned.drop_duplicates()
rows_after = df_cleaned.shape[0]
print(f"\n‚úÖ Removed {rows_before - rows_after} duplicate rows.")

# ---- 5.2 CLEAN TEXT COLUMNS (SPACES + UPPER CASE) ----
text_cols = df_cleaned.select_dtypes(include="object").columns
print("\nüìù Cleaning text columns:", list(text_cols))

for col in text_cols:
    # ŸÜÿ≠ŸàŸÑ ÿßŸÑŸÜÿµŸàÿµ ŸÑÿ≥ŸÄÿ™ÿ±ŸäŸÜÿ¨ Ÿà ŸÜÿ¥ŸäŸÑ ÿßŸÑŸÖÿ≥ÿßŸÅÿßÿ™ ÿßŸÑÿ≤ÿßŸäÿØÿ©
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip()
    df_cleaned[col] = df_cleaned[col].str.replace(r"\s+", " ", regex=True)
    # ŸÜÿÆŸÑŸä ÿßŸÑŸÜÿµŸàÿµ ŸÉŸÑŸáÿß UPPER CASE
    df_cleaned[col] = df_cleaned[col].str.upper()
    # ÿ£Ÿä ŸÜÿµ ŸÅÿßÿ∂Ÿä ŸÜÿÆŸÑŸäŸá NaN ÿπŸÑÿ¥ÿßŸÜ Ÿäÿ™ÿ¥ÿßŸÑ ŸÖÿπ ÿßŸÑŸÄ dropna
    df_cleaned[col] = df_cleaned[col].replace("", np.nan)

# ---- 5.3 REMOVE ROWS WITH MISSING VALUES ----
nulls_before = df_cleaned.isnull().sum().sum()
df_cleaned = df_cleaned.dropna()
nulls_after = df_cleaned.isnull().sum().sum()
print(f"\n‚úÖ Removed {nulls_before - nulls_after} missing values (rows containing NaN).")

# ---- 5.4 REMOVE OUTLIERS USING IQR (NUMERIC COLUMNS) ----
numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns
outliers_removed_total = 0

for col in numeric_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    before = df_cleaned.shape[0]
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower) & (df_cleaned[col] <= upper)]
    after = df_cleaned.shape[0]
    outliers_removed = before - after
    outliers_removed_total += outliers_removed

print(f"\n‚úÖ Removed approximately {outliers_removed_total} outlier rows (IQR method).")

# ==============================
# 6) SUMMARY AFTER CLEANING
# ==============================
print("\nüìå CLEANED dataset shape:", df_cleaned.shape)
print("\nüîπ Missing Values per Column (AFTER):")
print(df_cleaned.isnull().sum())
print("\nüîπ Number of duplicate rows (AFTER):",
      df_cleaned.duplicated().sum())

print("\nüìä Preview of CLEANED dataset:")
print(df_cleaned.head())

# ==============================
# 7) STATISTICAL SUMMARY AFTER CLEANING
# ==============================
print("\nüìà Starting statistical analysis of the CLEANED dataset...\n")

print("Dataset Info (AFTER cleaning):")
print(df_cleaned.info())
print("\n" + "="*70 + "\n")

print("Descriptive Statistics for Numeric Columns (AFTER cleaning):")
display(df_cleaned.describe().T)

numeric_df = df_cleaned.select_dtypes(include=[np.number])

stats = pd.DataFrame({
    "Mean": numeric_df.mean(),
    "Median": numeric_df.median(),
    "Std Dev": numeric_df.std(),
    "Variance": numeric_df.var(),
    "Min": numeric_df.min(),
    "Max": numeric_df.max()
})
print("\nüìä Detailed Numeric Summary:")
display(stats)

print("\nüîó Correlation Matrix (AFTER cleaning):")
display(numeric_df.corr())

print("\n‚ö† Checking for remaining extreme values (Z-score > 3):")
for col in numeric_df.columns:
    z_scores = np.abs((numeric_df[col] - numeric_df[col].mean()) / numeric_df[col].std())
    extreme_values = (z_scores > 3).sum()
    if extreme_values > 0:
        print(f"Column '{col}' still has {extreme_values} extreme values (|Z| > 3)")
    else:
        print(f"Column '{col}' looks normal ")

print("\n‚úÖ Statistical analysis completed successfully!")

# ==============================
# 8) VISUAL COMPARISON BEFORE vs AFTER
# ==============================
def plot_nulls_comparison(df_before, df_after):
    nulls_before = df_before.isnull().sum().sum()
    nulls_after = df_after.isnull().sum().sum()

    plt.figure(figsize=(6, 4))
    sns.barplot(
        x=["Before Cleaning", "After Cleaning"],
        y=[nulls_before, nulls_after],
        hue=["Before", "After"],
        palette=["red", "green"],
        legend=False
    )
    plt.title("Null Values Before vs After Cleaning")
    plt.ylabel("Count")
    plt.show()


def plot_duplicates_comparison(df_before, df_after):
    duplicates_before = df_before.duplicated().sum()
    duplicates_after = df_after.duplicated().sum()

    plt.figure(figsize=(6, 4))
    sns.barplot(
        x=["Before Cleaning", "After Cleaning"],
        y=[duplicates_before, duplicates_after],
        hue=["Before", "After"],
        palette=["orange", "green"],
        legend=False
    )
    plt.title("Duplicate Rows Before vs After Cleaning")
    plt.ylabel("Count")
    plt.show()


def plot_outliers_histogram_and_boxplot(df_before, df_after):
    numeric_cols = df_before.select_dtypes(include='number').columns

    for col in numeric_cols:
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))

        # Histogram Comparison
        sns.histplot(df_before[col], color='red',
                     label='Before Cleaning', kde=True,
                     bins=30, alpha=0.6, ax=axes[0])
        sns.histplot(df_after[col], color='green',
                     label='After Cleaning', kde=True,
                     bins=30, alpha=0.6, ax=axes[0])
        axes[0].set_title(f"Distribution of '{col}' Before & After Cleaning")
        axes[0].set_xlabel(col)
        axes[0].set_ylabel("Frequency")
        axes[0].legend()

        # Boxplot Comparison
        data = [df_before[col].dropna(), df_after[col].dropna()]
        axes[1].boxplot(
            data,
            patch_artist=True,
            boxprops=dict(facecolor='lightcoral'),
            medianprops=dict(color='black')
        )
        axes[1].set_xticklabels(["Before", "After"])
        axes[1].set_title(f"Boxplot of '{col}' Before & After Cleaning")
        axes[1].set_ylabel(col)

        plt.tight_layout()
        plt.show()


def plot_correlation_heatmaps(df_before, df_after):
    numeric_before = df_before.select_dtypes(include='number')
    numeric_after = df_after.select_dtypes(include='number')

    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    sns.heatmap(numeric_before.corr(), annot=True, fmt=".2f",
                cmap="coolwarm", ax=axes[0])
    axes[0].set_title("Correlation Heatmap - BEFORE Cleaning")

    sns.heatmap(numeric_after.corr(), annot=True, fmt=".2f",
                cmap="coolwarm", ax=axes[1])
    axes[1].set_title("Correlation Heatmap - AFTER Cleaning")

    plt.tight_layout()
    plt.show()


print("\nüìä Generating visual comparison BEFORE vs AFTER cleaning...")

plot_nulls_comparison(df_original, df_cleaned)
plot_duplicates_comparison(df_original, df_cleaned)
plot_outliers_histogram_and_boxplot(df_original, df_cleaned)
plot_correlation_heatmaps(df_original, df_cleaned)

# ==============================
# 9) SAVE CLEANED DATA
# ==============================
df_cleaned.to_csv("cleaned_data.csv", index=False)
print("\nüíæ File 'cleaned_data.csv' has been saved in the current working folder.")
```
